```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt  
```

```python
# Gradient Descent for Multiple Variables
def gradient_descent(X, y, theta, alpha, num_iters):
    m = len(y)
    J_history = np.zeros((num_iters, 1)).reshape(-1,1)
    for iter in range(num_iters):
        h = np.dot(X,theta)
        err = h - y
        theta_change = (alpha/m)*np.dot(X.T,err)
        theta = theta - theta_change
        J_history[iter] = (1/(2*m))*np.sum((np.dot(X,theta) - y)**2)
    return theta, J_history
```

```python
# Load data
in_data = pd.read_csv('data2.txt', header=None)
in_data = np.array(in_data)
x = in_data[:, 0:2]
y = in_data[:, 2:3]
m = len(y)

# Scale features and set them to zero mean
mu = np.mean(x, axis=0)
sigma = np.std(x, axis=0)
x = (x - mu)/sigma

# Add intercept term to X
x = np.insert(x,0,1,axis=1) # Add intercept term to X
theta = np.zeros((np.size(x,1), 1))
iterations = 400
alpha = 0.01

# Run gradient descent
[theta_new, j_his] = gradient_descent(x, y, theta, alpha, iterations)
```

```python
# Plot the convergence of the cost function
plt.plot(np.linspace(1,len(j_his),len(j_his)), j_his)
plt.xlabel('Number of iterations')
plt.ylabel('Cost J')
plt.show()
```

```python
# Make predictions
x_predict = [1650, 3]
price = theta_new[0] + (x_predict[0] - mu[0])*theta_new[1]/sigma[0] + (x_predict[1] - mu[1])*theta_new[2]/sigma[1]
print(price[0])
```

```python
# Normal Equation for Multiple Variables
in_data = pd.read_csv('data2.txt', header=None)
in_data = np.array(in_data)
x = in_data[:, 0:2]
y = in_data[:, 2:3]
m = len(y)
x = np.insert(x, 0, 1, axis=1)
a1 = np.linalg.inv(np.dot(x.T,x))
a2 = np.dot(x.T,y)
theta_n = np.dot(a1,a2)
x_pred = [1650, 3]
price = theta_n[0] + x_pred[0]*theta_n[1] + x_pred[1]*theta_n[2]
print(price[0])
```
