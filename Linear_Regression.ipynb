```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt  
```

```python
# Gradient Descent Function
def gradient_descent(X, y, theta, alpha, num_iters):
    m = len(y)
    J_history = np.zeros((num_iters, 1)).reshape(-1,1)

    for iter in range(num_iters):
        h = np.dot(X,theta)
        err = h - y
        theta_change = (alpha/m)*np.dot(X.T,err)
        theta = theta - theta_change
        J_history[iter] = (1/(2*m))*np.sum((np.dot(X,theta) - y.T)**2)
    return theta, J_history
```

```python
# Load Data
in_data = pd.read_csv('data1.txt', header=None)
x = np.array(in_data[0]).reshape(-1,1)
y = np.array(in_data[1]).reshape(-1,1)
m = len(y)
plt.plot(x, y, 'r+')
plt.show()
```

```python
# Add a column of ones to x (intercept term)
x_new = np.insert(x,0,1,axis=1)
theta = np.zeros((2, 1))
iterations = 1500
alpha = 0.01
```

```python
# Run Gradient Descent
[theta_new,b] = gradient_descent(x_new, y, theta, alpha, iterations)
y_new = np.dot(x_new,theta_new)
plt.plot(x,y_new)
plt.show()
```

```python
# Make a prediction for population = 35,000
predict = np.dot([1, 3.5],theta_new)
print(predict[0]*10000)
```

```python
# Visualizing J(theta0, theta1)
theta0_vals = np.linspace(-10, 10, 100)
theta1_vals = np.linspace(-1, 4, 100)
xx, yy = np.meshgrid(theta0_vals, theta1_vals)
j_vals = np.zeros((len(theta0_vals), len(theta1_vals)))
for i in range(len(theta0_vals)):
	for j in range(len(theta1_vals)):
		t = np.array((theta0_vals[i], theta1_vals[j])).reshape(2,1)
		j_vals[i][j] = (1/(2*m))*np.sum((x_new.dot(t) - y)**2)
```

```python
# 3D Surface Plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(xx, yy, j_vals, cmap='viridis', edgecolor='none')
#fig.colorbar(surf, ax=ax, shrink=0.5, aspect=15)
ax.set_xlabel('theta_0')
ax.set_ylabel('theta_1')
ax.set_zlabel('J_vals')
ax.set_title('3D Surface Plot Example')
plt.show()
```

```python
# Contour Plot
plt.figure
plt.contour(theta0_vals, theta1_vals, j_vals, levels=35, cmap='viridis')
plt.colorbar(label="da")
plt.title("J_val")
plt.xlabel('theta_0')
plt.ylabel('theta_1')
plt.plot(theta[0], theta[1], 'rx')
plt.show()
```
